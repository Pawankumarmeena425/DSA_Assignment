```python
# Normal Curve
#seaborn package
import seaborn as sns
flights = sns.load_dataset("flights")
flights.head()
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```python
 # B. DENSITY AND CONTOUR PLOTS
iris = sns.load_dataset("iris")
sns.kdeplot(data=iris)
```

```python
 # CORRELATION AND SCATTER PLOTS
# CORRELATION AND SCATTER PLOTS
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation visualized using a heatmap
df_corr = sns.load_dataset("titanic")
correlation_matrix = df_corr.corr()
ax = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()

# Scatter plots of categorical variable
df_scatter = sns.load_dataset("titanic")
sns.catplot(data=df_scatter, x="age", y="fare", hue="class", kind="scatter")
plt.title("Scatter Plot of Age vs. Fare with Class")
plt.show()
```

```python
 # D. HISTOGRAMS
#histogram of datafra,e
df = sns.load_dataset("titanic")
sns.histplot(data=df, x="age")
```

```python
# THREE DIMENSIONAL PLOTTING
#3d plotting using ploty package
import plotly as px
df = sns.load_dataset("iris")
px.scatter_3d(df, x="PetalLengthCm", y="PetalWidthCm", z="SepalWidthCm",
              size="SepalLengthCm",
              color="Species", color_discrete_map = {"Joly": "blue", "Bergeron": "violet","Coderre":"pink"})
```
