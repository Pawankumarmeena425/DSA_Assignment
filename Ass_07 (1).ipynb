```python
# Install Basemap toolkit


# Import necessary libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# Create a Basemap instance
m = Basemap(projection='mill',llcrnrlat=-60,urcrnrlat=90,llcrnrlon=-180,urcrnrlon=180,resolution='c')

# Plotting a point on the map
m.plot(0, 0, marker='o', color='red', markersize=10)

# Adding labels and title
m.drawcountries()
m.drawcoastlines()
m.drawmapboundary(fill_color='aqua')
plt.title('Sample Map with Basemap')

# Display the map
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution=None,width=8E6, height=8E6,lat_0=45, lon_0=-100,)
m.etopo(scale=0.5, alpha=0.5)
# Map (long, lat) to (x, y) for plotting
x, y = m(-122.3, 47.6)
plt.plot(x, y, 'ok', markersize=5)
plt.text(x, y, ' Seattle', fontsize=12);
```
